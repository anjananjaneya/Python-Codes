**Merging the data file and country code  file**

```python
final_df=pd.merge(data,data_country,on= 'Country Code',how='left')
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names= final_df.Country.value_counts().index
country_names
```

```python
country_values =final_df.Country.value_counts().values
country_values
```

```python
#pie chart
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

**Observation:** Zomato maximun records or Transaction are from India After that USA and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'reting count'})
ratings
```

**Obsarvation :**


*  When rating is between 4.5 to 4.9 - Excellent
*  When rating is between 4.0 to 4.4 - Very good
*  When rating is between 3.5 to 3.9 - Good
*  When rating is between 2.5 to 3.4 - Avarage
*  When rating is between 0.1 to 2.4 - Poor




```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sn.barplot(x="Aggregate rating", y="reting count",hue= "Rating color",data=ratings,palette =["gray","red","orange","yellow","green","green"])
```

**Obsarvation:**


*  Not Rated count is High
*  Maximum Number of Rating are between 3.0 to 3.9
